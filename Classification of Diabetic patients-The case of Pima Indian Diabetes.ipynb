```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
dia_data = pd.read_csv("diabetes.csv")
```

```python
dia_data
```

```python
dia_data.describe()
```

```python
dia_data['Diabetes'].replace(['pos', 'neg'], [1, 0], inplace= True) #replaces pos i.e. positive as 1 and neg i.e. negative as 0
```

```python
dia_data
```

```python
x = dia_data.iloc[:,0:7]
```

```python
x 
```

```python
y = dia_data.iloc[:,8]
```

```python
y #output 1 is pos and 0 is neg
```

```python
from sklearn import preprocessing
minmax=preprocessing.MinMaxScaler(feature_range=(0,1)) #Normalizing data to reshape
minmax.fit(x).transform(x) #reshaping the data into the dataframe
```

```python
from sklearn import neighbors, cross_validation
from sklearn.cross_validation import train_test_split
x_train, x_test, y_train,  y_test = train_test_split(x, y, test_size=0.2)
```

```python
x_train
```

```python
y_train = y_train.astype('int')
y_train
```

```python
x_test
```

```python
y_test = y_test.astype('int')
y_test
```

# kNN Classifier

```python
#fitting kNN classifier to training set
clf = neighbors.KNeighborsClassifier()
clf.fit(x_train, y_train)
```

```python
#Predicting on test data
y_pred = clf.predict(x_test)
```

```python
print("Actual diabetes: ", y_test.values)
```

```python
print("\nPredicted diabetes: ", y_pred)
```

```python
#Accuracy score on Test and Train
from sklearn.metrics import accuracy_score, recall_score, roc_auc_score, confusion_matrix

print("\nAccuracy score: %f" %(accuracy_score(y_test,y_pred) * 100))
print("Recall score : %f" %(recall_score(y_test, y_pred) * 100))
print("ROC score : %f\n" %(roc_auc_score(y_test, y_pred) * 100))
print(confusion_matrix(y_test, y_pred)) 
```

```python
#Now predicting the probability of each data point
probas = clf.predict_proba(x_test)
```

```python
plt.figure(dpi=150)
plt.hist(probas, bins=20)
plt.title('Classification Probabilities')
plt.xlabel('Probability')
plt.ylabel('Number of Instances of Diabetes')
plt.xlim([0.5, 1.0])
plt.legend(y_test)
plt.show()
```

```python
x_train_std = minmax.fit_transform(x_train)
x_test_std = minmax.transform(x_test)
```

```python
#Evaluation Metrics
from sklearn.model_selection import cross_val_score, cross_val_predict
```

```python
#Print cross validation score for kNN
clf_acc = cross_val_score(clf, x_train_std, y_train, cv=3, scoring="accuracy", n_jobs=-1)
```

```python
clf_proba = cross_val_predict(clf, x_train_std, y_train, cv=3, method='predict_proba') #cross validation prediction probability
clf_scores = clf_proba[:, 1] # get the score of the positive class
```

```python
clf_acc
```

```python
from sklearn.metrics import roc_auc_score,roc_curve
def ROC_curve(title, y_train, scores, label=None):
    # calculate the ROC score
    fpr, tpr, thresholds = roc_curve(y_train, scores)
    print('AUC Score ({}): {:.2f} '.format(title, roc_auc_score(y_train, scores)))
    
    # plot the ROC curve
    plt.figure(figsize=(8,6))
    plt.plot(fpr, tpr, linewidth=2, label=label, color='b')
    plt.xlabel('False Positive Rate', fontsize=16)
    plt.ylabel('True Positive Rate', fontsize=16)
    plt.title('ROC Curve: {}'.format(title), fontsize=16)
    plt.show()
```
